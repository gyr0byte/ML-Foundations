# NumPy Boolean Arrays

Boolean arrays are arrays whose elements are either `True` or `False`. In NumPy, they are commonly created by comparing numeric arrays with conditions such as greater than, less than, or equality.

A boolean array is often called a **mask** because it acts like a filter over another array:
- `True` means keep/select the element.
- `False` means ignore/remove the element.

## Why Boolean Arrays Are Useful

- They allow condition-based selection without manual loops.
- They make data filtering readable and concise.
- They are fast because operations are performed in optimized NumPy routines.

## Typical Operations with Boolean Arrays

- **Filtering**: Select elements that satisfy a condition.
- **Counting conditions**: Measure how many elements meet a rule.
- **Combining conditions**: Build complex filters using logical operations (AND, OR, NOT).
- **Conditional replacement**: Update values where a condition is true.

## Logical Combination of Masks

Multiple conditions can be merged into one mask to express complex criteria. This is commonly used in data cleaning and preprocessing, where you want to isolate invalid, missing, or out-of-range values.

## Important Concept

Boolean indexing usually returns a new array containing matching elements. This behavior is different from simple slicing, which often returns a view. Understanding this distinction helps avoid confusion when modifying data.

## Practical Perspective

Boolean arrays are one of NumPy's most important tools for analytical workflows. They let you describe *what* data you want in a declarative way, making array manipulation more expressive, maintainable, and efficient.
