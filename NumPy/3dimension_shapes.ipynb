# Dimensions and Shapes in NumPy

In NumPy, the **dimension** of an array tells you how many axes it has, while the **shape** shows the size of the array along each axis.

- A 1D array has one axis, so its shape looks like `(n,)`.
- A 2D array has rows and columns, so its shape looks like `(rows, columns)`.
- A 3D array adds another axis, often thought of as layers, rows, and columns.

You can use `ndim` to check how many dimensions an array has, and `shape` to see its full structure.

```python
import numpy as np
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
```

**Shape**

```python
A.shape
```
