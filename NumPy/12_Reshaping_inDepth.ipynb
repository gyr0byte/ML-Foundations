```python
arr1 = np.arange(6).reshape(2,3)
arr1
```

```python
arr2 = np.arange(12, 18).reshape(2, 3)
arr2
```

**Stacking** -> Stacking is the process of joining arrays along a new axis.

```python
np.hstack((arr1, arr2)) # Horizontal stacking
```

```python
np.vstack((arr1, arr2)) # Vertical stacking
```
