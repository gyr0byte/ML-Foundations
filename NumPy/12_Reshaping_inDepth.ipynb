# Array Reshaping Operations

This notebook explores in-depth array reshaping and manipulation techniques:

- **ndim**: Check the number of dimensions of an array
- **ravel()**: Flatten multidimensional arrays into 1D
- **transpose()**: Swap rows and columns (or reorder axes)
- **hstack()**: Stack arrays horizontally (column-wise)
- **vstack()**: Stack arrays vertically (row-wise)
- **hsplit()**: Split an array horizontally into sub-arrays
- **vsplit()**: Split an array vertically into sub-arrays

These operations are essential for reshaping data to fit different computational needs.

```python
import numpy as np
arr = np.array([[6, 7], [8, 9], [10, 11]])
```

```python
arr
```

```python
arr.ndim
```

**Ravel** -> Converts higher dimensional (2D, 3D, 4D...) arrays into 1D array

```python
arr.ravel()
```

**Transpose**

```python
arr.transpose()
```

```python
arr1 = np.arange(6).reshape(2,3)
arr1
```

```python
arr2 = np.arange(12, 18).reshape(2, 3)
arr2
```

**Stacking** -> Stacking is the process of joining arrays along a new axis.

```python
np.hstack((arr1, arr2)) # Horizontal stacking
```

```python
np.vstack((arr1, arr2)) # Vertical stacking
```

**Splitting** -> Splitting is the process of dividing an array into multiple sub-arrays.

```python
np.hsplit(arr1, 3) # Horizontal splitting
```

```python
np.vsplit(arr1, 2) # Vertical splitting
```
