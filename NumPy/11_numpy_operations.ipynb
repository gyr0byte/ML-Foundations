```python
import numpy as np
arr1 = np.array([1, 2, 3, 4, 5, 6])
arr2 = np.array([4, 5, 6, 7, 8, 9])
```

**Subtraction**

```python
arr1 - arr2
```

**Multiplication**

```python
arr1 * arr2 # vector multiplication (element-wise)
```

```python
arr1 * 2 # scalar multiplication
```

**Boolean Operations**

```python
arr1 > 3
```

**Dot Product**

```python
arr3 = np.arange(6).reshape(2, 3)
arr4 = np.arange(6, 12).reshape(3, 2)
arr3.dot(arr4)
```

**Maximum and Minimum**

```python
arr4.max(), arr4.min()
```

**Axis**


```python
arr4.min(axis=0), arr4.min(axis=1), arr4.max(axis=0), arr4.max(axis=1)
```

**Sin**

```python
np.sin(arr4)
```

**Mean and Median**

```python
arr4.mean(), np.median(arr4) , np.mean(arr4)
```

**Exponential and Logarithm**

```python
np.exp(arr4), np.log(arr4)
```
