<h2>Nepal Cricket Score Analyzer</h2>

- Create a NumPy array of 10 match scores
- Find highest score, lowest score, average
- Find matches where Nepal scored above 250
- Normalize all scores between 0 and 1
- Find the index of the highest scoring match
- Calculate standard deviation of scores

```python
import numpy as np
score = np.random.randint(150,320,10)
score
```

**Highest** and **Lowest** score

```python
score.max(), score.min()
```

**Average**

```python
score.mean()
```

**Score** above 250

```python
score[score[:] > 250]
```

**Normalize** all score between 0 and 1

```python
score_norm = (score - score.min()) / (score.max() - score.min())
score_norm
```
