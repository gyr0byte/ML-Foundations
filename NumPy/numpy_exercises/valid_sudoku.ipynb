```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

for **Rows**

```python
b = np.sum(s, axis=1)
for i in b:
    if i != 45:
        print("Sudoku is not valid!")
        break
else:
    print("It is valid for rows!")
```

for **Column**

```python
c = np.sum(s, axis=0)
for i in c:
    if i != 45:
        print("Sudoku is not valid!1")
        break
else:
    print("Sudoku is valid for columns!")
```

For 3x3 **boxes**

```python
valid = True
for i in range(0, 9, 3):
    for j in range(0, 9, 3):
        n = s[i:i+3, j:j+3]
        if n.sum() != 45:
            valid = False
            break
    if not valid:
        break
if valid:
    print("Sudoku is valid for boxes!")
else:
    print("Not a valid sudoku!")
```
