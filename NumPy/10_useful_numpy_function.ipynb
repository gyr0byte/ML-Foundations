# Summary of Functions Below

This notebook demonstrates common NumPy utilities used in the cells below:

- `import numpy as np`: import alias for NumPy.
- Random generators: `np.random.random(size)`, `np.random.normal(size)`, `np.random.rand(...)` — uniform, normal, and multi-dimensional random samples.
- `np.arange(start, stop, step)`: create ranges (examples: `np.arange(10)`, `np.arange(5, 10)`, `np.arange(0, 1, 0.1)`).
- `reshape()`: change array dimensions (e.g., `np.arange(10).reshape(2, 5)`).
- `np.linspace(start, stop, num, endpoint=True)`: evenly spaced values.
- Creation helpers: `np.zeros()`, `np.ones()`, `np.empty()` with optional `dtype` to control memory.
- Identity matrices: `np.identity(n)`, `np.eye(N, M, k)`.
- Simple Python example: string indexing (`"Hello world"[6]`).

Use these functions to create, sample, reshape, and inspect arrays for numerical workflows.
