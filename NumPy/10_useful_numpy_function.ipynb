# Useful NumPy Functions

NumPy provides a comprehensive set of functions for numerical computing and array manipulation. Here are some of the most commonly used ones:

## Array Creation & Manipulation
- **np.array()**: Create arrays from lists or tuples
- **np.zeros(), np.ones()**: Create arrays filled with zeros or ones
- **np.arange(), np.linspace()**: Generate arrays with specific ranges or intervals
- **np.reshape(), np.flatten()**: Change array dimensions
- **np.concatenate(), np.stack()**: Combine multiple arrays

## Mathematical Operations
- **np.sum(), np.mean(), np.std()**: Aggregate functions
- **np.min(), np.max()**: Find extremes
- **np.dot(), np.matmul()**: Matrix operations
- **np.sqrt(), np.exp(), np.log()**: Element-wise math functions

## Searching & Sorting
- **np.where()**: Find elements meeting a condition
- **np.sort(), np.argsort()**: Sort arrays
- **np.unique()**: Find unique elements

## Linear Algebra
- **np.linalg.inv()**: Matrix inversion
- **np.linalg.det()**: Determinant
- **np.linalg.eig()**: Eigenvalues and eigenvectors
