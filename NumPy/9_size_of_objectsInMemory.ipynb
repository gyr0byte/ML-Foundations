# Size of Objects in Memory

Understanding memory usage is important for optimizing performance and managing resources efficiently. Different data types and structures consume different amounts of memory.

## Key Concepts
- **sys.getsizeof()**: Returns the memory size in bytes of a Python object
- **NumPy arrays**: More memory-efficient than Python lists for numerical data
- **Data types**: Different dtypes (int32, int64, float32, float64) consume different amounts of memory

## Memory Efficiency
NumPy arrays are typically more memory-efficient than native Python lists because:
- They store homogeneous data types
- They have less overhead per element
- They use contiguous memory blocks

Use `array.nbytes` to check total bytes used by NumPy array data.

```python
import numpy as np
import sys
```

**Int and Float**


```python
# An integer in python is > 24 bytes
sys.getsizeof(1)
```

```python
# longs are even longer
sys.getsizeof(10**100)
```

```python
# Numpy size is much smaller
np.dtype(int).itemsize
```

```python
np.dtype(float).itemsize
```

**List** are even larger

```python
sys.getsizeof([1])
```

```python
# an array of one element in Numpy
np.array([1]).nbytes
```

And **performance** is also important

```python
l = list(range(100000))
```

```python
a = np.arange(100000)
```

```python
%time np.sum(a ** 2)
```

```python
%time sum([x**2 for x in l])
```
