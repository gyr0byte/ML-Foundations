# NumPy — Statistics

NumPy provides a comprehensive suite of statistical functions for computing descriptive statistics, distributions, and aggregations on arrays. These functions are optimized for performance and work seamlessly on multi-dimensional data.

## Descriptive Statistics

**Mean & Median**: Measures of central tendency.
- `np.mean(arr)` — Average of all elements
- `np.median(arr)` — Middle value when sorted
- `np.average(arr, weights=w)` — Weighted average

**Variance & Standard Deviation**: Measures of spread.
- `np.var(arr)` — Variance (average squared deviation from mean)
- `np.std(arr)` — Standard deviation (square root of variance)
- Lower std indicates data clustered closely around the mean

**Extremes & Range**:
- `np.min(arr)`, `np.max(arr)` — Minimum and maximum values
- `np.ptp(arr)` — Peak-to-peak (max - min), a simple range measure
- `np.percentile(arr, q)` — Value at a given percentile (0–100)
- `np.quantile(arr, q)` — Value at a given quantile (0–1)

## Aggregation Functions

Most statistical functions support:
- **axis parameter**: Compute along specific dimensions (axis=0 for columns, axis=1 for rows)
- **keepdims parameter**: Preserve the reduced dimension as size 1 for broadcasting

Example: On a 2D array, `np.mean(arr, axis=0)` computes column-wise means.

## Statistical Distributions & Summaries

- `np.sum(arr)` — Total of all elements
- `np.prod(arr)` — Product of all elements
- `np.cumsum(arr)` — Cumulative sum (running total)
- `np.cumprod(arr)` — Cumulative product
- `np.diff(arr)` — Differences between consecutive elements

## Correlation & Covariance

- `np.cov(arr)` — Covariance matrix; measures how two variables vary together
- `np.corrcoef(arr)` — Correlation coefficient matrix; normalized covariance (−1 to 1 range)

## Key Concepts

**Axis notation**: For a 2D array, axis=0 refers to rows (vertical), axis=1 refers to columns (horizontal).

**Broadcasting**: Statistics functions automatically broadcast results, making it easy to combine scalar and array operations.

**NaN handling**: Use `np.nanmean()`, `np.nanstd()`, `np.nansum()`, etc. to ignore NaN values in computations.

**Data type matters**: Integer arrays may produce integer results; use `.astype(float)` or pass `dtype=float` to ensure float output when needed for accurate statistics.
