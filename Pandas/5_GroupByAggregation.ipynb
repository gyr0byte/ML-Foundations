# GroupBy and Aggregation in Pandas

This notebook demonstrates grouping and summarizing sales data using Pandas:

- Create a DataFrame with `Category`, `Store`, `Sales`, `Quantity`, and `Date`
- Group by a single column and sum sales:
  - `df.groupby('Category')['Sales'].sum()`
  - `df.groupby('Store')['Sales'].sum()`
- Group by multiple columns:
  - `df.groupby(['Store', 'Category'])['Sales'].sum()`
- Perform aggregation on a column:
  - `df['Sales'].mean()`
  - `df['Sales'].agg(['mean','sum','median','min','max'])`

These operations help summarize data and compare performance across groups.

```python
import numpy as np
import pandas as pd
```

**Group By**

```python
data = {
    'Category' : ['A','B','A','B','A','B','A','B'],
    'Store' : ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales' : [100,200,150,250,120,180,130,220],
    'Quantity' : [10,20,15,25,12,18,13,22],
    'Date' : pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)
```

```python
df
```

Group by category and calculate the sum of sales

```python
cat = df.groupby('Category')['Sales'].sum()
cat
```

Group by store and calculate the sum of sales

```python
st = df.groupby('Store')['Sales'].sum()
st
```

Group by multiple columns

```python
mul = df.groupby(['Store','Category'])['Sales'].sum()
mul
```
