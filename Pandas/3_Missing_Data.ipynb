```python
import numpy as np
import pandas as pd
```

**Finding Missing Data**

```python
data = {
    'A' : [1, 2, np.nan, 4, 5],
    'B' : [np.nan, 2, 3, 4, 5],
    'C' : [1, 2, 3, np.nan, np.nan],
    'D' : [1, np.nan, np.nan, np.nan, 5]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

counting null values in each columns

```python
df.isna().sum()
```

checking if a column contains null value or not

```python
df.isna().any()
```
