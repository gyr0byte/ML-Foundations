```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10, 20 ,30]
arr = np.array([10, 20, 30])
d = {1:10, 2:20, 3:30}
```

**Series Creation**

```python
pd.Series(my_list)
```

Setting up your own **Index**

```python
pd.Series(my_list, index=labels)
```

**Using np array**

```python
pd.Series(arr)
```

```python
pd.Series(arr, index=labels)
```

**Using Dictionary**

```python
pd.Series(d)
```
