# Pivot Tables and Cross Tabs

This notebook demonstrates reshaping data using pivot tables and cross tabs:

**Pivot Tables:**
- Create a DataFrame with `Date`, `Product`, `Region`, `Sales`, `Units`, and `Rep` columns
- Extract month and quarter from the date using `.dt` accessor
- `pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='median')`: Reshape data with rows as index, columns as headers, and median aggregation
- `pd.pivot_table(df, values=['Sales','Units'], index='Region', columns='Product')`: Pivot multiple value columns

**Cross Tabs:**
- `pd.crosstab(df['Region'], df['Product'])`: Create a frequency table of two categorical variables

Use these to summarize and reorganize data for reporting and analysis.

```python
import numpy as np
import pandas as pd
```

**Pivot Tables**

```python
data = {
    'Date' : pd.date_range(start='2023-01-01', periods=20),
    'Product' : ['A','B','C','D']*5,
    'Region' : ['East', 'West', 'North', 'South']*5,
    'Sales' : np.random.randint(100, 1000,20),
    'Units' : np.random.randint(10, 100, 20),
    'Rep' : ['John', 'Mary', 'Bob', 'Alice']*5
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quater'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='median')
```

```python
pivot2 = pd.pivot_table(df, values=['Sales','Units'], index='Region', columns='Product')
pivot2
```

**Cross Tabs**

```python
pd.crosstab(df['Region'],df['Product'])
```
