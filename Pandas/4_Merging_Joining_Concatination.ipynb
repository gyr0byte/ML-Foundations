```python
import numpy as np
import pandas as pd
```

**Merging two dataframes**

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'employee_name':['John','Jane','Jack','Jill','Jerry'],
    'department' :['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_id':[1,2,3,6,7],
    'salary':[50000,60000,70000,80000,90000],
    'bonus':[5000,6000,7000,8000,9000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries) # automatically does on the basis of common column
```

on keyword

```python
pd.merge(employees,salaries,on='employee_id') # on -> provided the common column
```

how keyword

```python
pd.merge(employees,salaries,on='employee_id',how='inner') # how is inner by default
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer') # outer -> all the records from both the dataframes, 
                                                          # if there is no match then it will fill with NaN
```

```python
pd.merge(employees,salaries,on='employee_id',how='left') # left -> all the records from the left dataframe, 
                                                          # if there is no match then it will fill with NaN
```

```python
pd.merge(employees,salaries,on='employee_id',how='right') # right -> all the records from the right dataframe,
                                                          # if there is no match then it will fill with NaN
```

**Concatination of two dataframes**

```python
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2'],
    'B' : ['B0','B1','B2'],
    'C' : ['C0','C1','C2'],
})

df2 = pd.DataFrame({
    'A' : ['A3','A4','A5'],
    'B' : ['B3','B4','B5'],
    'C' : ['C3','C4','C5'],
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # on the basis of columns, axis = 0 by default
```

```python
pd.concat([df1,df2], axis=1) # on the basis of rows, axis = 1
```

**Joining two dataframes**

```python
df1 = pd.DataFrame({
    'name' : ['John','Jane','Jack'],
}, index=[1,2,3])

df2 = pd.DataFrame({
    'score' : [90,80,70],
}, index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```
